# Q4

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("Cars Company.csv")  # Make sure the file is in the same directory

# Step 2: Group by 'Year' and calculate the average MSRP
msrp_by_year = df.groupby("Year")["MSRP"].mean().round()

# Step 3: Plot the data using a line plot
plt.figure(figsize=(12, 6))
plt.plot(msrp_by_year.index, msrp_by_year.values, marker='o', linestyle='-', color='orange')

# Step 4: Add title and labels
plt.title("Average MSRP Over the Years")
plt.xlabel("Year")
plt.ylabel("Average MSRP (USD)")
plt.grid(True, linestyle='--', alpha=0.6)

# Step 5: Display the plot
plt.tight_layout()
plt.show()
```

# Q5

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("Cars Company.csv")  # CSV file must be in the same directory

# Step 2: Calculate Average MPG = (city mpg + highway MPG) / 2
df["Average MPG"] = (df["city mpg"] + df["highway MPG"]) / 2

# Step 3: Group by 'Year' and calculate average Engine HP and average MPG
engine_hp_by_year = df.groupby("Year")["Engine HP"].mean().round()
avg_mpg_by_year = df.groupby("Year")["Average MPG"].mean().round()

# Step 4: Create the plot
plt.figure(figsize=(12, 6))

# Plot for Engine HP
plt.plot(engine_hp_by_year.index, engine_hp_by_year.values, marker='o', linestyle='-', label='Average Engine HP', color='blue')

# Plot for Average MPG
plt.plot(avg_mpg_by_year.index, avg_mpg_by_year.values, marker='s', linestyle='--', label='Average MPG', color='green')

# Step 5: Add labels, title, and legend
plt.title("Average Engine HP and MPG Over the Years")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Step 6: Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("Cars Company.csv")  # Make sure your CSV file is in the same folder

# Step 2: Filter only Compact, Midsize, Large vehicles
vehicle_sizes = ['Compact', 'Midsize', 'Large']
filtered_df = df[df['Vehicle Size'].isin(vehicle_sizes)]

# Step 3: Count and calculate percentage
size_counts = filtered_df['Vehicle Size'].value_counts()
total_cars = len(df)
percentage_sizes = (size_counts / total_cars) * 100

# Step 4: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentage_sizes, 
        labels=percentage_sizes.index, 
        autopct='%1.1f%%', 
        startangle=140,
        colors=['#66b3ff', '#99ff99', '#ffcc99'])

plt.title("Percentage of Cars by Size: Compact, Midsize, or Large")
plt.axis('equal')  # Ensures pie is circular
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("Cars Company.csv")  # Make sure the CSV is in the same directory

# Step 2: Count the number of vehicles by style
style_counts = df["Vehicle Style"].value_counts()

# Step 3: Plot vertical bar chart
plt.figure(figsize=(12, 6))
plt.bar(style_counts.index, style_counts.values, color='skyblue')

# Step 4: Labels and title
plt.xlabel("Vehicle Style")
plt.ylabel("Number of Cars Sold")
plt.title("Most Sold Vehicle Styles")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 5: Show plot
plt.tight_layout()
plt.show()
```
